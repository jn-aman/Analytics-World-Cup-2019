```python
counts_all=dict()
def word_count(str1):
    counts = {
        'india': 0,
  'pakistan': 0,
  'sri_lanka': 0,
  'england': 0,
  'afganistan': 0,
  'australia': 0,
  'bangladesh': 0,
  'new_zealand': 0,
  'south_africa': 0,
    "west_indies": 0}
    
    words = vocab

    for word in vocab:
        if word=='sri lanka':
            counts['sri_lanka']=str1.count(word)
            
        elif word=='new zealand':
            counts['new_zealand']=str1.count(word)
            
        elif word=='south africa':
            counts['south_africa']=str1.count(word)
            
        elif word=='west indies':
            counts['west_indies']=str1.count(word)
        else:

            counts[word]=str1.count(word)
        

    return counts
```

```python
for i in range(len(tweets)):
    counts_all[i]=word_count(tweets[i]["text"])
```
